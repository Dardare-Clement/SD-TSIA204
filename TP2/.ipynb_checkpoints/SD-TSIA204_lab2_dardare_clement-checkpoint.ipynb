## <u>Question 2)</u>

```python
def forward_variable_selection(X, Y):
    r = Y.copy()
    n, p = X.shape                # Dimension of X
    A = [k for k in range(p)]     # Creating lists for the variables
    S = []                          # For the selected covariates
    order = []                      # For orders of the variable selectioned
    for i in range(p):
        Ts = []  # list for the test stats
        indexes = []  # list for indexes
        for k in A:
            if (k not in order):

                Xk_tild = X[:, k].reshape(-1, 1)
                Xk = np.hstack((np.ones(Xk_tild.shape), Xk_tild))
                # We need to compute the OLS, s and sigma to compute Tk.

                # We compute OLS, sigma, the gram matrix for s and s:
                model = LinearRegression()
                model = model.fit(Xk, Y)

                model_coef = model.coef_.T
                Y_pred = model.predict(Xk)

                # We compute the residual and all we need to compute the test
                # statistics
                r = Y - Y_pred

                sigma = np.sqrt(1 / (n - 2)) * np.linalg.norm(r)**2

                gram = (1 / n) * Xk.T @ Xk

                e1 = np.array([[0, 1]]).T

                s = np.sqrt(e1.T @ np.linalg.inv(gram) @ e1)

                # Calculates the test statistics

                Tk = np.sqrt(n) * np.abs(model_coef) / (s * sigma)
                Ts.append(Tk[0][1])
                indexes.append(k)

        # we search the index of argmax(T)

        ind = Ts.index(max(Ts))

        Ks = indexes[ind]
        order.append(Ks)

        # We compute the p_value
        # Calculates the p-value of the estimator
        p_value = 2 * (1 - stat.norm.cdf(Ts[ind]))
        S.append([Xk_tild, p_value])
    return (S, order)
```

```python
S, order = forward_variable_selection(X_train, Y_train)
```

# Ridge

## <u>Question 5)</u>
---

```python
def ridge(X, Y, alpha):
    n, p = X.shape

    # We compute the minimizer of the ridge estimator thanks to the formula
    A = np.identity(p)
    Theta_Rdg = np.linalg.inv(X.T @ X + n * alpha * A) @ X.T @ Y

    return (Theta_Rdg)
```

```python
# Create alphas values
alpha_values = np.logspace(-9, 2, num=100, base=10)

# To store coefficients
thetas = []
for alpha in alpha_values:
    thetas.append(ridge(X_train, Y_train, alpha))
thetas = np.array(thetas)

# Plot the coefficients against log(alpha)
fig, ax = plt.subplots(figsize=(10, 7))
ax.plot(alpha_values, thetas)
ax.set_xscale('log')
ax.set_xlabel('alpha')
ax.set_ylabel('Coefficients')
ax.set_title('Ridge coefficients as a function of the penalty')
plt.show()
```

```python
# We compute the MSE for each value of alpha
train_mse = []  # store MSE of train data
test_mse = []  # store MSE of test data

for alpha in alpha_values:
    # compute coefficents thanks to ridge
    estimator = ridge(X_train, Y_train, alpha)
    Y_train_pred = X_train @ estimator  # compute train pred
    Y_test_pred = X_test @ estimator  # compute test pred
    train_mse.append(
        mean_squared_error(
            Y_train,
            Y_train_pred))  # add MSE to train_mse
    test_mse.append(
        mean_squared_error(
            Y_test,
            Y_test_pred))  # add MSE to test_mse

# find the best mse of both list
argmin1 = np.argmin(train_mse)
argmin2 = np.argmin(test_mse)

print(f'best alpha value is={alpha_values[argmin2]}')

# Plot the MSE against penalty
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(
    alpha_values,
    train_mse,
    '-D',
    markevery=[argmin1],
    label='Training set')
ax.plot(alpha_values, test_mse, '-D', markevery=[argmin2], label='Test set')
ax.set_xscale('log')
ax.set_xlabel('alpha')
ax.set_ylabel('MSE')
ax.set_title('MSE as a function of the penalty')
ax.legend()
plt.show()

print(alpha_values[argmin2])
```

```python
# c)
# compute and storer2 score of the estimator (alpha=10**-9)
estimator = ridge(X_train, Y_train, 10**(-9))
r2 = r2_score(Y_test, X_test @ estimator)

df_coef = pd.concat([df_coef, pd.DataFrame(
    {'method': ["ridge"], 'R2': [r2]})], ignore_index=True)
print(f'coefficient of determination R2={r2:.4f}')
```

# Crossvalidation, Lasso and elastic net

## <u>Question 6)</u>

```python
# a)

# We compute the MSE for each value of alpha
train_mse = []  # store MSE of train data
test_mse = []  # store MSE of test data
alpha_values2 = np.logspace(-5, -2, num=100, base=10)

# We search coefficients different from 0
nb_coef_non_null = []
for pen in alpha_values2:
    # Apply model lasso with alpha
    model = Lasso(alpha=pen, tol=0.05, max_iter=10000)
    model = model.fit(X_train, Y_train)
    coefs = model.coef_
    Y_train_pred = X_train @ coefs
    Y_test_pred = X_test @ coefs

    # Compute the MSE
    train_mse.append(mean_squared_error(Y_train, Y_train_pred))
    test_mse.append(mean_squared_error(Y_test, Y_test_pred))
    nb_coef = 0

    # Compute the number of coefficient different from 0
    for coef in coefs:
        if coef != 0:
            nb_coef += 1
    nb_coef_non_null.append(nb_coef)
```

```python
# Plot the number of coefficients different from 0 as a function of the penalty
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(alpha_values2, nb_coef_non_null)
ax.set_xscale('log')
ax.set_xlabel('penalty')
ax.set_ylabel('Number of coefficients different from 0')
ax.set_title(
    'Number of coefficients different from 0 as a function of the penalty')
plt.show()
```

```python
# c)
# We find the best alpha
argmin1 = np.argmin(train_mse)
argmin2 = np.argmin(test_mse)

# Plot the MSE against alpha for both train and test sets
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(
    alpha_values2,
    train_mse,
    '-D',
    markevery=[argmin1],
    label='Training set')
ax.plot(alpha_values2, test_mse, '-D', markevery=[argmin2], label='Test set')
ax.set_xscale('log')
ax.set_xlabel('alpha')
ax.set_ylabel('MSE')
ax.set_title('MSE as a function of the penalty')
ax.legend()
plt.show()

test_mse[argmin2]
```

```python
# d)
model = Lasso(alpha=alpha_values2[argmin2], tol=0.05, max_iter=10000)
model = model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)

R2_Las = r2_score(Y_test, Y_pred)

df_coef = pd.concat([df_coef, pd.DataFrame(
    {'method': ["Lasso"], 'R2': [R2_Las]})], ignore_index=True)
print(f'coefficient of determination R2={R2_Las:.4f}')
```

# Bootstrap

## <u>Question 8)</u>

```python
X40 = X.iloc[:, 39]  # store the 40th variable

# Run OLS on the data
model = LinearRegression()
model.fit(np.array(X40).reshape(-1, 1), Y)
coef = model.coef_
intercept = model.intercept_

# Compute the regression line
xs = np.linspace(np.min(X40), np.max(X40), 100)
ys = intercept + coef * xs

# plot the dataset ant the regression line
plt.figure(figsize=(10, 7))
plt.scatter(X40, Y, label='Y data')
plt.plot(xs, ys, label='regression line', color='red')
plt.xlabel('X40 values')
plt.ylabel('Y values')
plt.title('dataset and regression line in function of the 40th variable')
plt.legend()
plt.show()
```

```python
X40 = X.iloc[:, 39]  #store X40
X40 = X40.to_numpy()
# Set the number of bootstrap samples
n_bootstrap = 50

# Initialize matrix to store coefficients of the regression models
coef_matrix = np.zeros((n_bootstrap, 2))

n = len(X40)
bootstrapped_preds = []

#Generate 50 bootstrap samples
plt.figure(figsize=(12,7))

for i in range(n_bootstrap):
    indices = np.random.choice(n, n, replace=True) # randomly select indices with replacement
    bootstrap_sample = X40[indices] # select the samples corresponding to the indices
    bootsrap_Y = Y.to_numpy().flatten()[indices]
    
    # Run OLS on the data
    model = LinearRegression()
    model.fit(np.array(bootstrap_sample).reshape(-1, 1), bootsrap_Y)
    coef = model.coef_
    intercept = model.intercept_
    Y_pred = np.polyval([coef, intercept], X40)
    # Compute the regression line
    xs = np.linspace(np.min(bootstrap_sample), np.max(bootstrap_sample), 100)
    ys = intercept + coef * xs
    
    # Plot regrsion line
    plt.plot(xs, ys, alpha=0.4)
    
    # calculate prediction interval
    Y_pred_sample = np.polyval(coef, bootstrap_sample)
    sigma2 = np.sum((Y_pred_sample - bootsrap_Y) ** 2) / (len(bootstrap_sample) - 2)
    sigma = np.sqrt(sigma2)
    t = stat.t.ppf(0.975, len(bootstrap_sample) - 2)
    Y_err = t * sigma * np.sqrt(1 + 1 / len(X40) + (X40 - np.mean(bootstrap_sample)) ** 2 / np.sum((bootstrap_sample - np.mean(bootstrap_sample)) ** 2))
    bootstrapped_preds.append(Y_pred)
    plt.fill_between(X40, Y_pred - Y_err, Y_pred + Y_err, alpha=0.05, color='orange')
    
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('50 estimated regression lines')
plt.show()
```

# PCA

## <u>Question 9)</u>

```python
# a)

# Compute covariance matrix
cov_matrix = np.cov(X_train.T)

# Compute Singular Value Decomposition of the covariance matrix
U, s, V = np.linalg.svd(cov_matrix)

# Plot Heatmap of the covariance matrix
plt.figure(figsize=(15, 8))
sns.heatmap(cov_matrix)
plt.title("Covariance Matrix Heatmap")
plt.show()
```

```python
# c)
# Calculate the amount variance explained
explained_variance = (s**2) / np.sum(s**2)

# Plot the amount of variance explained by the first k principal components
plt.figure(figsize=(12, 5))
plt.plot(np.arange(2, len(explained_variance) + 1),
         np.cumsum(explained_variance[1:]))
plt.title('amount of variance explained')
plt.xlabel('Number of components')
plt.ylabel
```

```python
# d)
# Compute th PCs
PC = np.dot(X_train, U)

# Plot the first and second PCs
plt.figure(figsize=(12, 7))
plt.scatter(PC[:, 0], PC[:, 1], c=Y_train)
plt.title("First and Second PCs")
plt.xlabel("First PC")
plt.ylabel("Second PC")
plt.colorbar()
plt.show()
```

```python
#e)
# Calculate the projected data for all values of k in 2..p
p = len(s)  
X_train_proj = X_train @ V[:p].T   #Compute X_train in the new base
X_test_proj = X_test @ V[:p].T     #Compute X_test in the new base
k_range = range(2, p+1)

# Run OLS on the projected data using k components
scores = []    #store the scores
for k in k_range:
    #Selected the k components for X_train and X_test
    X_train_proj_selected = X_train_proj[:, :k]   
    X_test_proj_selected = X_test_proj[:, :k]
    
    #Run OLS
    model = LinearRegression()
    model.fit(X_train_proj_selected, Y_train)
    Y_test_pred = model.predict(X_test_proj_selected)
    
    #Store R2
    scores.append(r2_score(Y_test, Y_test_pred))

#find the best R2
R2_pca = np.max(scores)                               
df_coef = pd.concat([df_coef, pd.DataFrame({'method': ["PCA"], 'R2': [R2_pca]})], ignore_index=True)
print(f'coefficient of determination R2={R2_pca:.4f}')
```
